```python
def create_boxplot(data, axs, ix, iy, tit):
    bp = axs[ix,iy].boxplot(data, patch_artist = True, vert = 0, notch = 0)

    colors = ['blue','red','green','yellow']
    
    for patch, color in zip(bp['boxes'], colors):
        patch.set_facecolor(color)

    # changing color and linewidth of
    # medians
    for median in bp['medians']:
        median.set(color ='black',
                linewidth = 3)
        x, y = median.get_xydata()[1] # top of median line
        axs[ix,iy].text(x, y+0.05, '%.1f' % x,
            horizontalalignment='center') # draw above, centered
    
    # changing style of fliers
    for flier in bp['fliers']:
        flier.set(marker ='D',
                color ='#e7298a',
                alpha = 0.25)

    # changing color and linewidth of
    # caps
    for cap in bp['caps']:
        cap.set(color = '#8B008B',
                linewidth = 2)

    # changing color and linewidth of
    # whiskers
    for whisker in bp['whiskers']:
        whisker.set(color ='#8B008B',
                    linewidth = 1.5,
                    linestyle =":")

    # Setting limits and labels
    #axs[ix,iy].set_xlim([-5,300])
    axs[ix,iy].set_yticklabels(['Sinoidal', 'Bezier12','Cycloid','Simple'])
    axs[ix,iy].set_title(tit)
    axs[ix,iy].set_xlabel("Force (N)")

    
    return bp
```

```python
fig,axs = plt.subplots(2 ,2 ,figsize =(13, 10))
 
bp_fr=create_boxplot(data_FR,axs,0,0,"FR foot forces")
bp_fl=create_boxplot(data_FL,axs,0,1,"FL foot forces")
bp_br=create_boxplot(data_BR,axs,1,0,"BR foot forces")
bp_bl=create_boxplot(data_BL,axs,1,1,"BL foot forces")

# show plot
plt.show(bp_fr)  
```
