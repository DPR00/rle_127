```python
import numpy as np
import matplotlib.pyplot as plt
```

## Trayectorias Sinusoidales

```python
th = np.linspace(0, 2*np.pi,100)
x = np.zeros(len(th))
z = np.zeros(len(th))
dx = 0.08; dz = 0.1
for i in range(len(th)):
    x[i] = dx*np.cos(th[i])
    #res = np.floor(th[i]/(2*np.pi))
    th[i] %= 2*np.pi #- 2*np.pi*res
    if th[i] > np.pi:
        z[i] = 0
    else:
        z[i] = dz*np.sin(th[i])

plt.plot(x,z,'g')
plt.show()
```

## Trayectoria Cicloide

```python
th = np.linspace(0*np.pi, 4*np.pi,100)
x = np.zeros(len(th))
z = np.zeros(len(th))
dx = 0.08; dz = 0.1
for i in range(len(th)):
    #res = np.floor(th[i]/(2*np.pi))
    th[i] %= 4*np.pi #- 2*np.pi*res
    if th[i] <= 2*np.pi:
        x[i] = dx*(th[i]-np.sin(th[i]))/(2*np.pi)
        if th[i] >= np.pi:    
            z[i] = 2*(dz -dz*(2*th[i]-np.sin(2*th[i]))/(4*np.pi))
        else:
            z[i] = 2*dz*(2*th[i]-np.sin(2*th[i]))/(4*np.pi)
    else:
        th[i] %= 2*np.pi
        x[i] = dx*(th[i]-np.sin(th[i]))/(2*np.pi)
        z[i] = 0
th1 = np.linspace(0*np.pi, 20*np.pi,100)

plt.plot(x,z,'g')
plt.show()
```
