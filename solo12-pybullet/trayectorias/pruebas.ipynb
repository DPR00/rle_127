```python
import numpy as np
import matplotlib.pyplot as plt
```

## Trayectorias Sinusoidales

```python
th = np.linspace(0, 2*np.pi,100)
x = np.zeros(len(th))
z = np.zeros(len(th))
dx = 0.08; dz = 0.1
for i in range(len(th)):
    x[i] = dx*np.cos(th[i])
    #res = np.floor(th[i]/(2*np.pi))
    th[i] %= 2*np.pi #- 2*np.pi*res
    if th[i] > np.pi:
        z[i] = 0
    else:
        z[i] = dz*np.sin(th[i])

plt.plot(th,x,'g')
plt.show()
```

## Trayectorias simples

```python
th = np.linspace(0, 0.8*np.pi,500)
x = np.zeros(len(th))
z = np.zeros(len(th))
dx = 0.08; dz = 0.1; dt = th[1]-th[0]
for i in range(len(th)):
    # x[i] = dx*np.cos(th[i])
    #res = np.floor(th[i]/(2*np.pi))
    th[i] %= 2*np.pi #- 2*np.pi*res
    if th[i] <= np.pi:
        x[i] = 2*th[i]*dx/np.pi -dx
        if th[i] <= np.pi/2:
            z[i] = 2*th[i]*dz/np.pi
        else:
            z[i] = 2*dz-2*th[i]*dz/np.pi   
    else:
        x[i] = -2*th[i]*dx/np.pi + 3*dx
        z[i] = 0

plt.plot(th,x,'g')
plt.show()
```

```python
th = np.linspace(0*np.pi, 1.2*np.pi,500)
th1 = np.linspace(0*np.pi, 4*np.pi, len(th))
x = np.zeros(len(th))
z = np.zeros(len(th))
dx = 0.08; dz = 0.1
for i in range(len(th)):
    #res = np.floor(th[i]/(2*np.pi))
    th[i] %= 4*np.pi #- 2*np.pi*res
    if th[i] <= 2*np.pi:
        x[i] = 2*dx*(th[i]-np.sin(th[i]))/(2*np.pi)-dx
        if th[i] >= np.pi:    
            z[i] = 2*(dz -dz*(2*th[i]-np.sin(2*th[i]))/(4*np.pi))
        else:
            z[i] = 2*dz*(2*th[i]-np.sin(2*th[i]))/(4*np.pi)
    else:
        #th[i] %= 2*np.pi
        x[i] = dx-2*dx*(th[i]%(2*np.pi)-np.sin(th[i]%(2*np.pi)))/(2*np.pi)
        z[i] = 0

plt.plot(x,z,'g')
plt.xlabel("Distancia en el eje x (m)")
plt.ylabel("Distancia en el eje z (m)")
plt.title("Trayectoria de la pata del robot")
plt.show()
```
